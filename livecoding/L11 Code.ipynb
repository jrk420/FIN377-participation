```python
#Q3 - JPY per EUR is "EURJPY" 2 years ago
baseurl = 'https://api.exchangeratesapi.io/2018-01-02?base=EUR&symbols=JPY'
r = requests.get(baseurl)
pd.DataFrame(r.json())
```

```python
#Q3 - JPY per EUR is "EURJPY" 2 years ago
baseurl = 'https://api.exchangeratesapi.io/2018-01-02?'
params = {'base':'EUR','symbols':'JPY'}
r = requests.get(baseurl,params=params)
pd.DataFrame(r.json())
```

```python
#Q4: Get a time series of EURJPY from 2018 through 2019

baseurl = 'https://api.exchangeratesapi.io/history?'
params = {'base':'EUR','symbols':'JPY','start_at':'2018-01-01','end_at':'2019-12-31'}
r = requests.get(baseurl,params=params)
df = pd.DataFrame(r.json())
df['EURJPY'] = df['rates'].apply(lambda x: x.get('JPY'))
df.describe()
```
